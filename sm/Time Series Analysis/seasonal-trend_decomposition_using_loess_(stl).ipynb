# [Seasonal-Trend decomposition using LOESS (STL)](https://www.statsmodels.org/dev/examples/notebooks/generated/stl_decomposition.html)

本笔记本说明了如何使用将 `STL` 时间序列分解为三个组成部分：趋势、季节 和残差。`STL` 使用 LOESS（局部估计散点图平滑）来提取三个分量的平滑估计。`STL` 关键输入是：
- `season` - 季节的长度更加平滑。一定是奇数。
- `trend` - 趋势的长度比较平滑，通常为 `season` 150% 左右。必须是奇数且大于 `season`。
- `low_pass` - 低通估计窗口的长度，通常是大于数据周期性的最小奇数。

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
sns.set_style('darkgrid')
plt.rc('figure', figsize=(16, 12))
plt.rc('font', size=13)
```

## `robust` 设置

设置 `robust` 使用数据相关的加权函数，该函数在估计 LOESS 时重新加权数据（使用 LOWESS 也是如此）。使用稳健估计使模型能够容忍底部图上可见的较大误差。

```python
from statsmodels.datasets import elec_equip as ds
elec_equip = ds.load().data.iloc[:, 0]
```

接下来，我们估计具有和不具有鲁棒权重的模型。这种差异很小，并且在 2008 年金融危机期间最为明显。非稳健估计对所有观测值赋予相同的权重，因此平均产生较小的误差。权重在 0 和 1 之间变化。

```python
def add_stl_plot(fig, res, legend):
    axs = fig.get_axes()
    comps = ["trend", "seasonal", "resid"]
    for ax, comp in zip(axs[1:], comps):
        series = getattr(res, comp)
        if comp == "resid":
            ax.plot(series, marker="o", linestyle="none")
        else:
            ax.plot(series)
            if comp == "trend":
                ax.legend(legend, frameon=False)
```

```python
stl = STL(elec_equip, period=12, robust=True)
res_robust = stl.fit()
fig = res_robust.plot()
res_non_robust = STL(elec_equip, period=12, robust=False).fit()
add_stl_plot(fig, res_non_robust, ['Robust', 'Non-robust'])
```

```python
fig = plt.figure(figsize=(16, 5))
lines = plt.plot(res_robust.weights, marker='o', linestyle='none')
ax = plt.gca()
xlim =  ax.set_xlim(elec_equip.index[0], elec_equip.index[-1])
```

## LOESS degree

默认配置使用常数和趋势来估计 LOESS 模型。通过将 `COMPONENT_deg` 设置为 0，可以将其更改为仅包含常量。除了 2008 年金融危机期间的趋势外，这里的 `degree` 几乎没有产生任何区别。

```python
stl = STL(
    elec_equip, period=12, seasonal_deg=0, trend_deg=0, low_pass_deg=0, robust=True
)
res_deg_0 = stl.fit()
fig = res_robust.plot()
add_stl_plot(fig, res_deg_0, ["Degree 1", "Degree 0"])
```

## 性能

可以使用三个选项来降低 `STL` 分解的计算成本：
- `seasonal_jump`
- `trend_jump`
- `low_pass_jump`

当这些值非零时，仅在 `COMPONENT_jump` 观测值中估计分量 `COMPONENT` 的 LOESS，并且在点之间使用线性插值。这些值通常不应分别超过季节性、趋势或 `low_pass` 大小的 10-20%。

下面的示例展示了如何使用具有低频余弦趋势和正弦季节性模式的模拟数据将计算成本降低 15 倍。

```python
import numpy as np

rs = np.random.RandomState(0xA4FD94BC)
tau = 2000
t = np.arange(tau)
period = int(0.05 * tau)
seasonal = period + ((period % 2) == 0)  # ensure odd
e = 0.25 * rs.standard_normal(tau)
y = np.cos(t / tau * 2 * np.pi) + 0.25 * np.sin(t / period * 2 * np.pi) + e
plt.plot(y)
plt.title("Simulated Data")
xlim = plt.gca().set_xlim(0, tau)
```

首先，估计基线模型，所有跳跃都等于 1。

```python
mod = STL(y, period=period, seasonal=seasonal)
%timeit mod.fit()
res = mod.fit()
fig = res.plot(observed=False, resid=False)
```

跳跃都设置为窗口长度的 15%。有限的线性插值对模型的拟合影响不大。

```python
low_pass_jump = seasonal_jump = int(0.15 * (period + 1))
# The length of the trend smoother, usually around 150% of season.
trend_jump = int(0.15 * 1.5 * (period + 1))
mod = STL(
    y,
    period=period,
    seasonal=seasonal,
    seasonal_jump=seasonal_jump,
    trend_jump=trend_jump,
    low_pass_jump=low_pass_jump,
)
%timeit mod.fit()
res = mod.fit()
fig = res.plot(observed=False, resid=False)
```

使用 STL 预测

`STLForecast` 简化了使用 STL 消除季节性因素，然后使用标准时间序列模型预测趋势和周期性成分的过程。

在这里，我们使用 STL 来处理季节性，然后使用 ARIMA(1,1,0) 对去季节数据进行建模。季节性成分是从查找完整周期中预测的，其中
$$E[S_{T+h}|\mathcal{F}_T]=\hat{S}_{T-k}$$

式中，$k= m - h + m \lfloor \frac{h-1}{m} \rfloor$。

预测会自动将季节性成分预测添加到 ARIMA 预测中。

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.forecasting.stl import STLForecast

elec_equip.index.freq = elec_equip.index.inferred_freq
stlf = STLForecast(elec_equip, ARIMA, model_kwargs=dict(order=(1, 1, 0), trend="t"))
stlf_res = stlf.fit()

forecast = stlf_res.forecast(24)
plt.plot(elec_equip)
plt.plot(forecast)
plt.show()
```

`summary` 包含有关时间序列模型和 STL 分解的信息。
